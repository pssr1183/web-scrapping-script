```python
class node:
    def __init__(self,val):
        self.left= None
        self.right=None
        self.val= val

def bstrange(root,l,h):

    if root is None:
        return
    if root.val>=l and root.val<=h:
        print(root.val,end=" ")
        count=count+1
    bstrange(root.left,l,h)
    bstrange(root.right,l,h)

root= node(5)
root.left = node(22)
root.right = node(12)
root.left.left = node(1)
root.left.right= node(3)
root.right.left = node(9)
root.right.right= node(21)

bstrange(root,1,10)
global count
# print(count)
```

```python
import requests, smtplib
from bs4 import BeautifulSoup

class Notification:

  def __init__(self, url, title):
    self.url = url
    self.title = title

  def check_for_new_notifications(self):
    """Checks if there are any new notifications on the given URL."""
    r = requests.get(self.url)
    soup = BeautifulSoup(r.content, "html.parser")
    headings = soup.find_all("li")
    new_notifications = []
    for heading in headings:
      title = str.strip(heading.text)
      if title not in self.title:
        new_notifications.append(title)
    self.title = new_notifications

  def send_email(self, recipient):
    """Sends an email to the given recipient with the list of new notifications."""
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.ehlo()
    server.starttls()
    server.ehlo()
    server.login('poornasai62@gmail.com', '8919504106')
    subject = 'New notifications from GVPCE'
    body = 'Here are the new notifications from GVPCE: \n' + '\n'.join(self.title)
    message = 'Subject: {}\n\n{}'.format(subject, body)
    server.sendmail(
      '20131a1210@gvpce.ac.in',
      recipient,
           message
    )
    server.quit()


def main():
  general_notification = Notification('https://www.gvpce.ac.in/General%20Notifications.html', [])
  exam_notification = Notification('https://www.gvpce.ac.in/examnotifications.html', [])
  result_notification = Notification('https://www.gvpce.ac.in/result.html', [])
  time_table_notification = Notification('https://www.gvpce.ac.in/timetables.html', [])

  general_notification.check_for_new_notifications()
  exam_notification.check_for_new_notifications()
  result_notification.check_for_new_notifications()
  time_table_notification.check_for_new_notifications()

  if general_notification.title or exam_notification.title or result_notification.title or time_table_notification.title:
    general_notification.send_email('poornasai62@gmail.com')

if __name__ == '__main__':
  main()
```
